```python
import collections.abc
collections.Mapping = collections.abc.Mapping

from experta import *

#create Shopping Assistant class thats derived from knowledge engine
class ShoppingAssistant(KnowledgeEngine):

    # Create rules for product recommendations
    @Rule(Fact(category="electronics"), Fact(price_range="medium"), Fact(features="battery_life"))
    def recommend_laptop(self):
        print("Recommendation: A laptop witha long battery life.")

    @Rule(Fact(category="electronics"), Fact(price_range="low"), Fact(features="waterresistant"))
    def recommend_smartphone(self):
        print("Recommendation: A waterresistant smartphone with a low price.")

    @Rule(Fact(category="electronics"), Fact(price="low"),Fact(  features="battery life"))
    def recommend_earbuds(self):
        print("Recommended Product: Budget earbuds with decent battery life.")

    


    # Handle missing inputs or for error handling
    @Rule(NOT(Fact(category=W())), salience=-1)
    # salience is used to set the priority of the rule
    
    def ask_category(self):
        print("Error: Please specify a category (eg, electronics, clothing).")

    @Rule(NOT(Fact(price_range=W())), salience=-1)
    def ask_price(self):
        print("Error: Please specify a price range (low, medium, high).")

    @Rule(NOT(Fact(features=W())), salience=-1)
    def ask_features(self):
        print("Error: Please specify a key feature (eg, waterproof, battery_life).")

    

# Function to collect user input and run the assistant
def get_recommendations():
    # Create an instance of the ShoppingAssistant class
    engine = ShoppingAssistant()

    #Reset the engine for  new inputs
    engine.reset()
  
    print("\nWelcome to the Online Shopping Assistant!")
    while True:
        print("\nEnter your preferences for a product recommendation:")

        # Get user inputs
        category = input("Category (electronics): ").strip().lower()
        #The input is declared as a fact in the system
        engine.declare(Fact(category=category))

        price_range = input("Price range (low, medium, high): ").strip().lower()
        engine.declare(Fact(price_range=price_range))

        features = input("Key features (e.g., waterresistant, battery_life): ").strip().lower()
        engine.declare(Fact(features=features))

        

        # Run the engine 
        print("\nProcessing your preferences...")
        engine.run()

        #  another recommendation
        another = input("\nWould you like another recommendation? (yes/no): ").strip().lower()
        if another != "yes":
            print("Thank you for using the Online Shopping Assistant!")
            break


# Run the assistant
get_recommendations()
```
