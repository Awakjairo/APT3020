```python
import random
import matplotlib.pyplot as plt

#Math Tutor, solving simple math problems for primary school students
# Step 1: Define the Domain Model
problems = [
    {"question": "5 + 3 =", "answer": 8},
    {"question": "12 - 4 =", "answer": 8},
    {"question": "3 x 7 =", "answer": 21},
    {"question": "16 รท 4 =", "answer": 4},
    {"question": "8 x 6 =", "answer": 48},
    {"question": "36 รท 6 =", "answer": 6},
    {"question": "9 + 7 =", "answer": 16},
    {"question": "20 - 5 =", "answer": 15},
    {"question": "4 x 9 =", "answer": 36},
    {"question": "45 รท 9 =", "answer": 5}
]

# Step 2: tracking process of the student in student model
student_model = {
    'correct': 0,
    'incorrect': 0
}

def update_student_model(is_correct):
    if is_correct:
        student_model['correct'] += 1
    else:
        student_model['incorrect'] += 1

# Step 3: Define the Teaching Model (Adapt Difficulty)
def get_next_question():
    if student_model['correct'] > student_model['incorrect']:
        # Return harder questions if the student is doing well
        return random.choice([
            {"question": "8 x 6 =", "answer": 48},
            {"question": "36 รท 6 =", "answer": 6},
            {"question": "12 x 5 =", "answer": 60},
            {"question": "72 รท 8 =", "answer": 9}
        ])
    else:
        # Return easier questions if the student is struggling
        return random.choice(problems)

# Step 4: user interface
def ask_question():
    question = get_next_question()
    print(question["question"])
    answer = int(input("Your answer: "))

    if answer == question["answer"]:
        print(" Correct!")
        update_student_model(True)
    else:
        print(f" Incorrect! The correct answer is {question['answer']}.")
        update_student_model(False)

# Step 5: Run the ITS (ask 5 questions
)
for _ in range(5):
    ask_question()

# Step 6: Plot Student Performance
attempts = list(range(1, student_model['correct'] + student_model['incorrect'] + 1))
correct_answers = [1 if i <= student_model['correct'] else 0 for i in attempts]

plt.plot(attempts, correct_answers, marker='o')
plt.title('Student Performance Over Time')
plt.xlabel('Attempt Number')
plt.ylabel('Correct/Incorrect')
plt.yticks([0, 1], ['Incorrect', 'Correct'])
plt.show()

# Step 7: Provide Adaptive Feedback
def give_feedback():
    if student_model['correct'] > student_model['incorrect']:
        print("\n๐ Great job! Let's have you atteempt harder problems")
    else:
        print("\n๐ Keep practicing! You can consider attempting some easier questions first.")

give_feedback()
```
